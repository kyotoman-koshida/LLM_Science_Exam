```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
train_df = pd.read_csv("../input/train.csv")
my_data_df = pd.read_csv("../input/stem_dataset_5k_2023_08_19.csv")
master_data_df = pd.read_csv("../input/6000_train_examples.csv")
```

```python
plt.hist([len(str(x).split(" ")) for x in train_df.prompt], bins=30, alpha=0.7, density=True, label="train.csv")
plt.hist([len(str(x).split(" ")) for x in my_data_df.prompt], bins=30, alpha=0.7, density=True, label="my_data")
plt.hist([len(str(x).split(" ")) for x in master_data_df.prompt], bins=30, alpha=0.7, density=True, label="masters_data")

plt.title("prompt")
plt.legend()
```
